```python
from tensorflow.keras.applications.resnet_v2 import ResNet50V2
cnn = ResNet50V2(include_top=False, input_shape=(224, 224, 3))
cnn.summary()
```

```python
from tensorflow.keras.utils import plot_model
plot_model(cnn, to_file="resnet.png", show_shapes=True)
```

```python
# 1 * 1 加的時候通道樹要混合，要一樣
w = cnn.get_layer("conv2_block1_0_conv").get_weights()
w[0].shape
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Conv2D, Add, Input

i = Input(shape=(56, 56, 64))
# line 1
x1 = Conv2D(256, 1, padding="same")(i)
# line 2
x2 = Conv2D(64, 1, activation="relu")(i)
x2 = Conv2D(64, 3, activation="relu", padding="same")(x2)
x2 = Conv2D(256, 1)(x2)
out = Add()([x1, x2])
model= Model(inputs=i, outputs=out)
model.summary()
```
